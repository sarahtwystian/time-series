```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import ListedColormap
```

```python
data=pd.read_csv("archive/time_series_data_human_activities.csv")
```

```python
data.head()
```

```python
for i in ["Walking", "Jogging", "Upstairs", "Downstairs", "Sitting", "Standing"]:
    X=np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['x-axis'])
    pd.Series(np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['x-axis'])).plot()
    print(i, 'x-axis')
    plt.show()
    pd.Series(np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['y-axis'])).plot()
    print(i, 'y-axis')
    plt.show()
    pd.Series(np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['z-axis'])).plot()
    print(i, 'z-axis')
    plt.show()
```

```python
def convert_float(x):
    a=x.real
    b=x.imag
    val=a+b 
    return val
#reference:https://stackoverflow.com/questions/35423928/how-do-i-convert-a-complex-number
```

```python
plt.specgram(x)
```

```python
for i in ["Walking", "Jogging", "Upstairs", "Downstairs", "Sitting", "Standing"]:
    x=np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['x-axis'][:300])#the pattern of small subset of data seems clear
    y=np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['y-axis'][:300])
    z=np.fft.fft(data[(data['user']==3)&(data['activity']==i)]['z-axis'][:300])
    x=np.array([convert_float(ele) for ele in x])
    x_plt=plt.specgram(x)
    plt.title(i+" x_axis")
    plt.show()
    y=np.array([convert_float(ele) for ele in y])
    y_plt=plt.specgram(y)
    plt.title(i+" y_axis")
    plt.show()
    z=np.array([convert_float(ele) for ele in z])
    z_plt=plt.specgram(z)
    plt.title(i+" z_axis")
    plt.show()
    
```

```python
for i in ["Walking", "Jogging", "Upstairs", "Downstairs", "Sitting", "Standing"]:
    x=np.fft.fft(data[(data['user']==5)&(data['activity']==i)]['x-axis'][:100])#the pattern of small subset of data seems clear
    y=np.fft.fft(data[(data['user']==5)&(data['activity']==i)]['y-axis'][:100])
    z=np.fft.fft(data[(data['user']==5)&(data['activity']==i)]['z-axis'][:100])
    x=np.array([convert_float(ele) for ele in x])
    x_plt=plt.specgram(x)
    plt.title(i+" x_axis")
    plt.show()
    y=np.array([convert_float(ele) for ele in y])
    y_plt=plt.specgram(y)
    plt.title(i+" y_axis")
    plt.show()
    z=np.array([convert_float(ele) for ele in z])
    z_plt=plt.specgram(z)
    plt.title(i+" z_axis")
    plt.show()
```

```python
x.shape[0]
```

```python
#PCA
A=np.array([x,y,z]).T
print(A)
a=A-np.mean(x,axis=0)
cov_m=(a.T@a)/A.shape[0]
val,vec=np.linalg.eig(cov_m)
v=range(len(val))
plt.scatter(v,val)
```

```python
import plotly.express as px
from sklearn.decomposition import PCA

pca = PCA()
components = pca.fit_transform([x,y,z])
components
```

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(4)
    
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
